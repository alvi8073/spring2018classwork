# Finding word frequencies using dicts and lists

```python
aliceFile = open("alice.txt")

aliceText = aliceFile.read()
```

```python
print aliceText
```

```python
aliceWords = aliceText.split()
```

```python
print len(aliceWords)
```

```python
print aliceWords[0:10]
```

```python
wordFrequencies = {}

for word in aliceWords:

    if wordFrequencies.get(word.lower()) != None:
        
        wordFrequencies[word.lower()] += 1
    else:
        wordFrequencies[word.lower()] = 1
```

```python
print wordFrequencies
print len(wordFrequencies)
```

```python
print len(wordFrequencies)
```

```python
print type(wordFrequencies)
```

```python
sortedWords = sorted(wordFrequencies,key=wordFrequencies.get,reverse = True)
```

```python
print sortedWords
```

```python
print sortedWords[0:10]
```

```python
for word in sortedWords[0:10]:
    print word,wordFrequencies[word]
```

```python
wordCounts = wordFrequencies.values()
print wordCounts
```

```python
wordCounts.sort()
```

```python
print wordCounts
```

```python
print wordCounts[:-11:-1]
```

```python
for count in wordCounts[:-11:-1]:
    
    for key in wordFrequencies:
        if count == wordFrequencies[key]:
            print key
```
