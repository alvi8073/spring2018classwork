# Bringing it all together

```python
movieInfo = {}

movieFile = open("mldata/movies.csv")

movieFile.readline()

line = movieFile.readline()

while line != "":
    lineContents = line.strip().split(",")
    if len(lineContents) != 3:
        line = movieFile.readline()
        continue
        
    genres = lineContents[2].split("|")
    title = lineContents[1]
    
    #titleDateList = lineContents[1].split("(")
    
    #date = titleDateList[-1]
    #date = date.strip(")")
    
    movieInfo[lineContents[0]] = [title,genres]
    
    line = movieFile.readline()
    

    
```

```python
movieInfo['156387']
```

```python
movieInfo['40']
```

# Counting movies per genre

```python
genreCounts = {}

count = 0
for key in movieInfo:
    
    count += 1
    
    genres =  movieInfo[key][-1]
    
    for genre in genres:
        
        if genreCounts.get(genre) == None:
            genreCounts[genre] = 1
        else:
            genreCounts[genre] += 1
    
```

```python
print genreCounts
```

# Writing to a file

```python
writeFile = open("salaryReport.txt","w")

writeFile.write("id,deptid,salary\n")

for i in range(1,101):
    
    id = i
    deptid = id%5 + 1
    salary = i*deptid*10000
    
    writeFile.write(str(id) + "," + str(deptid) + "," + str(salary) + "\n")

writeFile.close()
```
